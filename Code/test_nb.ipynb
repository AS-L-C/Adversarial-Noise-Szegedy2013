```python
from torchvision.models import resnet18, ResNet18_Weights

model = resnet18(weights=ResNet18_Weights.DEFAULT)
model.eval()
# Initialize the Weight Transforms
weights = ResNet18_Weights.DEFAULT
preprocess = weights.transforms()


# If you want to save the model weights
# torch.save(model.state_dict(), 'pretrained_resnet18.pth')
```

```python
from datasets import load_dataset

dataset = load_dataset(
    "imagenet-1k",
    split="validation",
    streaming=True,
    trust_remote_code=True,
).take(2000)
```

```python
import torch

matches = torch.Tensor((2000))
for ind, img_dict in enumerate(dataset):
    # img_dict = next(iter(dataset))
    print(ind)
    img, label = img_dict["image"], img_dict["label"]
    img = img if img.mode == "RGB" else img.convert("RGB")
    prep_img = preprocess(img).unsqueeze(0)
    # prep_img = prep_img.repeat(1, 3, 1, 1) if prep_img.shape[1] == 1 else prep_img
    scores = model(prep_img).squeeze(0).softmax(0)
    pr_label = scores.argmax().item()
    print(label)
    print(pr_label)
    print(scores[pr_label])
    matches[ind] = label == pr_label
print(ind)
print(matches.mean())
```

```python
from torchvision.transforms.functional import to_pil_image


def to_pil(img_tensor, prep_fcn=preprocess):
    stds = torch.tensor(prep_fcn.std)
    means = torch.tensor(prep_fcn.mean)
    img = img_tensor * stds.view(3, 1, 1) + means.view(3, 1, 1)
    return to_pil_image(img)


aa = to_pil(prep_img.squeeze(0))
aa.show()
```
